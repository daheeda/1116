# 정규표현식

- 파이썬 뿐만 아니라 모든 언어 공통, 복잡한 문자열을 처리할때 사용하는 기법.

- 일정한 규칙(패턴)을 가진 문자열을 표현하는 방법

- 복잡한 문자열 속에서 특정한 규칙으로 된 문자열을 검색한 뒤 추출 및 변경할 때 사용

- 문자열이 정해진 규칙에 부합하지는 확인 시에도 사용

조건 - 123456-1234567 : 6자리-7자리 이어야 할 것. 아니라면 error 메세지
     - 1234567 : 뒷자리 중 맨 앞은 1,2,3,4 중 하나여야 할 것. 아니라면 error 메세지

```python
data = "choi 123456-1234567"
data1 = "choi 123456-0234567"
data2 = "choi 12346-0234567"
```

```python
blur(data)
```

```python
blur(data1)
```

```python
blur(data2)
```

```python
def blur(data):
    newdata=data.split(" ")
    fdata=newdata[1].split("-")
    f=fdata[1]
    if (len(fdata[0])==6)&(len(fdata[1])==7)&((f[0]=='1')|(f[0]=='2')|(f[0]=='3')| (f[0]=='4')):
        print(newdata[0]," ",fdata[0],"-","*******")   
    else:
        print("error")
```

```python
import re # 정규표현식 모듈
pat=re.compile("(\d{6})(-)\d{7}")
print(pat.sub("\g<1>-*******",data))
```

```python
p = re.compile('[a-z]+')  # a부터 z까지중
m= p.match('python')     # python 과 일치하는 문자 있는지
print(m)
```

```python
p = re.compile('[a-z]+')  
m= p.match('3 python')    # 3 이란 숫자는 일치 x
print(m)
```

```python
p = re.compile('[a-z]+')  
m= p.findall('very good 323 ㅇㅇㅇ aa') # list로 출력    
print(m)
```
